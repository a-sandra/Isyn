# Fundamental machine and beam parameters

```python
speedoflight = 299792458.0
electroncharge = 1.60218e-19 
eps0= 8.85419e-12
mu0 = 4.*math.pi*10**(-7)
Z0= 377.0
```

```python
circum = 1083.6
bpipe = 0.1
beta = 0.99
gamma0 = 5.0
g0 = 3.84
R = 172.46
h=5
omega0 =beta*speedoflight/R
omega1 = 2*math.pi*beta*speedoflight/circum
print "omega %f omeganew %f" % (omega0,omega1)
offsetphi = 0.0 #
initialDeltaPhi = 0.48
```

```python
nParticle = 1000000
particleArray=np.zeros(nParticle,dtype="float")

random.seed(3000)
numberofParticle = 2.0e13
charge = 1.0*electroncharge
chargePernParticle = numberofParticle/nParticle
nc = 1.0e9 #circum/(2*math.pi*bpipe*beta)*1000 # kind of cut off frequency
```

# Useful functions

```python
" Particle shape "
def s(x):
    return max(1-abs(x),0)

" Space Charge Impedance "
spaceChargeImpedance = lambda w: (2*math.pi)*w*mu0*circum/(4*math.pi*(beta*gamma0)**2)*g0

" Analytical formula for space charge voltage"
def scVoltage(x):
    return 3.0/4.0*(charge*g0*numberofParticle*h*h)/(eps0*gamma0**2*initialDeltaPhi*initialDeltaPhi*initialDeltaPhi*R)*(-h/R*x)
```

# Generate the distribution

```python
" Generate parabolic distribution "
for i in np.arange(0,nParticle):
    u = random.uniform(0,1)
    v = random.uniform(0,1)
    phi = offsetphi + initialDeltaPhi * math.sqrt(1.0 - math.pow(1.0 - u, 2.0/3.0))*math.cos(2*math.pi*v)
    particleArray[i] = - (phi-offsetphi)*R/h
```
